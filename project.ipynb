#### Exercise 11. Analyze the relation between the surface and the price of the houses (★★☆)

#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in Exercise 14) (★★★)

For the normalization method, you can use the one you consider; there is not a single correct answer to this question. Print the plot and write in the Markdown cell a brief analysis about the plot.

Hint: You can help yourself by reviewing the *multihist* demo of Matplotlib.

```python
# TODO

import pandas as pd
import matplotlib.pyplot as plt

# Filtrar los datos para las áreas deseadas
areas = ['Fuenlabrada', 'Leganés', 'Getafe', 'Alcorcón']
ds_filtered = ds[ds['level5'].isin(areas)]

# Normalizar el precio para cada área
for area in areas:
    ds_filtered.loc[ds_filtered['level5'] == area, 'price_normalized'] = (ds_filtered[ds_filtered['level5'] == area]['price'] - ds_filtered[ds_filtered['level5'] == area]['price'].min()) / (ds_filtered[ds_filtered['level5'] == area]['price'].max() - ds_filtered[ds_filtered['level5'] == area]['price'].min())

# Crear el histograma
fig, ax = plt.subplots()
for area in areas:
    ax.hist(ds_filtered[ds_filtered['level5'] == area]['price_normalized'], bins=20, alpha=0.5, label=area)

ax.set_xlabel('Precio Normalizado')
ax.set_ylabel('Frecuencia')
ax.set_title('Distribución de Precios Normalizados por Área')
ax.legend()
plt.show()
```

#### Exercise 19. What can you say about the price per square meter (price/m2) between the towns of "Getafe" and "Alcorcón"? You must use the subset obtained in Exercise 14 (★★☆)

Hint: Create a new column called `pps` (price per square meter) and then analyze the values.

```python
# TODO

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns  # Para utilizar el estilo de seaborn en los gráficos

# Filtrar los datos para Getafe y Alcorcón
areas = ['Getafe', 'Alcorcón']
ds_filtered = ds[ds['level5'].isin(areas)]

# Calcular el precio por metro cuadrado (PPS)
ds_filtered['pps'] = ds_filtered['price'] / ds_filtered['surface']

# Crear un box plot para comparar la distribución del PPS
plt.figure(figsize=(10, 6))
sns.boxplot(x='level5', y='pps', data=ds_filtered)
plt.title('Comparación del Precio por Metro Cuadrado en Getafe y Alcorcón')
plt.xlabel('Área')
plt.ylabel('Precio por Metro Cuadrado')
plt.show()

# Calcular el PPS medio para cada área
mean_pps_getafe = ds_filtered[ds_filtered['level5'] == 'Getafe']['pps'].mean()
mean_pps_alcorcon = ds_filtered[ds_filtered['level5'] == 'Alcorcón']['pps'].mean()

print(f"Precio medio por metro cuadrado en Getafe: {round(mean_pps_getafe,2)}")
print(f"Precio medio por metro cuadrado en Alcorcón: {round(mean_pps_alcorcon,2)}")
```

#### Exercise 20. Make the same plot for 4 different populations (level5 column) and rearrange them on the same graph. You must use the subset obtained in Exercise 14 (★★☆)
 
Hint: Make a scatter plot of each population using subplots.

```python
# TODO

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Filtrar los datos para las áreas deseadas
areas = ['Fuenlabrada', 'Leganés', 'Getafe', 'Alcorcón']
ds_filtered = ds[ds['level5'].isin(areas)]

# Calcular el precio por metro cuadrado (PPS)
ds_filtered['pps'] = ds_filtered['price'] / ds_filtered['surface']

# Crear subplots para comparar la distribución del PPS across areas
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))

for i, area in enumerate(areas):
    row = i // 2
    col = i % 2
    sns.scatterplot(x='surface', y='pps', data=ds_filtered[ds_filtered['level5'] == area], ax=axes[row, col])
    axes[row, col].set_title(f'Precio por Metro Cuadrado en {area}')
    axes[row, col].set_xlabel('Superficie')
    axes[row, col].set_ylabel('Precio por Metro Cuadrado')

plt.tight_layout()
plt.show()
```
